```python
from sklearn.model_selection import GridSearchCV

clf = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
grid_search.fit(x_train, y_train)
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
best_model = grid_search.best_estimator_
test_score = best_model.score(x_test, y_test)
print("Test Accuracy:", test_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Initialize and train the RandomForest classifier
rf = RandomForestClassifier(max_depth=10)
rf.fit(x_train, y_train)

# Predict on training and testing data
y_train_rf_predicted = rf.predict(x_train)
y_test_rf_predicted = rf.predict(x_test)

# Training Performance
print('Training Model Performance:')
print('Training Accuracy:', accuracy_score(y_train, y_train_rf_predicted))
print('Training Precision:', precision_score(y_train, y_train_rf_predicted)) 
print('Training Recall:', recall_score(y_train, y_train_rf_predicted))  
print('Training Confusion Matrix: ', confusion_matrix(y_train, y_train_rf_predicted))
print('Training Classification Report:\n', classification_report(y_train, y_train_rf_predicted))

# Testing Performance
print('\nTesting Model Performance:')
print('Testing Accuracy:', accuracy_score(y_test, y_test_rf_predicted))
print('Testing Precision:', precision_score(y_test, y_test_rf_predicted)) 
print('Testing Recall:', recall_score(y_test, y_test_rf_predicted))  
print('Testing Classification Report:\n', classification_report(y_test, y_test_rf_predicted))
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Initialize the XGBoost classifier
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

# Fit the model on training data
xgb_model.fit(x_train, y_train)

# Predict on training and testing data
y_train_xgb_predicted = xgb_model.predict(x_train)
y_test_xgb_predicted = xgb_model.predict(x_test)

# Training Performance
print("Training Model Performance:")
print("Training Accuracy:", accuracy_score(y_train, y_train_xgb_predicted))
print("Training Precision:", precision_score(y_train, y_train_xgb_predicted, average='binary')) 
print("Training Recall:", recall_score(y_train, y_train_xgb_predicted, average='binary'))  
print("Training Confusion Matrix:\n", confusion_matrix(y_train, y_train_xgb_predicted))
print("Training Classification Report:\n", classification_report(y_train, y_train_xgb_predicted))

# Testing Performance
print("\nTesting Model Performance:")
print("Testing Accuracy:", accuracy_score(y_test, y_test_xgb_predicted))
print("Testing Precision:", precision_score(y_test, y_test_xgb_predicted, average='binary'))
print("Testing Recall:", recall_score(y_test, y_test_xgb_predicted, average='binary'))
print("Testing Confusion Matrix:\n", confusion_matrix(y_test, y_test_xgb_predicted))
print("Testing Classification Report:\n", classification_report(y_test, y_test_xgb_predicted))
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
model1=LogisticRegression()
model2=DecisionTreeClassifier()
model3=SVC()
model4=RandomForestClassifier()

ensemble_classifier=VotingClassifier(
    estimators=[('lr',model1),('dt2',model2),('rf',model4)],voting='soft')

adaboost=AdaBoostClassifier(
    estimator=ensemble_classifier,
    learning_rate=1.0,
    random_state=40
)

adaboost.fit(x_train,y_train)
y_train_pred=adaboost.predict(x_train)

print("Model Performance")
print("Target Accuracy: ",accuracy_score(y_train,y_train_pred))
#print("Training precision: ",precision_score(y_train,y_train_pred))
#print("Training recall: ",recall_score(y_train,y_train_pred))
print(confusion_matrix(y_train,y_train_pred))
print(classification_report(y_train,y_train_pred))

y_test_pred=adaboost.predict(x_test)


print("Model Performance")
print("Testing Accuracy: ",accuracy_score(y_test,y_test_pred))
#print("Training precision: ",precision_score(y_train,y_train_pred))
#print("Training recall: ",recall_score(y_train,y_train_pred))
print(confusion_matrix(y_test,y_test_pred))
print(classification_report(y_test,y_test_pred))
```
